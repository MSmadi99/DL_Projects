```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def htan(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def der_htan(x):
    return 1 - htan(x)**2
```

```python
X_data = np.linspace(-6, 6, 100)
y_data = htan(X_data)
dy_data = der_htan(X_data)
```

```python
plt.plot(X_data, y_data, X_data, dy_data)
plt.title('htan Activation Function and Derivative') 
plt.legend(['htan', 'der_htanh'])
plt.grid()
plt.show()
```
