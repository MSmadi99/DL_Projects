```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
X_data = np.linspace(-10, 10, 100)
y_data = softmax(X_data)
```

```python
plt.plot(X_data, y_data)
plt.title('Softmax Activation Function')
plt.grid()
plt.show()
```
